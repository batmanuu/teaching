# 1 - Carrega os dados

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler

dados = []
timestamps = []

with np.load('C:\\Users\\selah\\OneDrive\\Emprego\\Docencia\\git\\teaching\\data-mining-applied-to-engineering\\3-resources\\dataset_shrink_transformed.npz') as data:
    for field in data.files:
        print(field)
        campo = data[field]
        #print(campo.shape)
        if(len(campo.shape) == 2):
            dados = campo
        else:
            timestamps = campo

print(timestamps.shape)
print(dados.shape)

# compute required values
scaler = StandardScaler()
model = scaler.fit(dados)
dados_normalizados = model.transform(dados)

for i in range(0, 10):
    print(i)
    plt.figure(figsize=(20,10))
    plt.plot(dados_normalizados[:, i])
    plt.show()
```

```python
threshold = 2
sensor_escolhido = dados_normalizados[:, 2]
anomalies = sensor_escolhido[np.abs(sensor_escolhido) > threshold]
print(len(anomalies))
indices = np.where(np.abs(sensor_escolhido) > threshold)[0]

```

```python
plt.figure(figsize=(20,10))
plt.plot(sensor_escolhido, label='Dados')
plt.scatter(indices, sensor_escolhido[indices], c='r', label='Anomalias')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
dados_temporarios = dados_normalizados[:, 1]
dados_reduzidos = np.mean(dados_temporarios.reshape(-1, 200), axis=1)
len(dados_reduzidos)
plt.figure(figsize=(20,10))
plt.plot(dados_reduzidos)
plt.show()
```
